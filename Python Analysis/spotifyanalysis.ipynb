Case Study: Analyzing Spotify Data

```python
# Importing Libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Uploading Data file

```python
# Uploading Data file
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# Count the number of missing values in each column of the DataFrame df_tracks
pd.isnull(df_tracks).sum()
```

```python
# Display information about the DataFrame df_tracks, including data types and memory usage
df_tracks.info()
```

# Ranging Popularity Ascending : 

```python
# Sort the DataFrame df_tracks by the 'popularity' column in ascending order
# Select the top 10 rows with the lowest popularity values
sorted_df = df_tracks.sort_values('popularity', ascending=True).head(10)

# Display the sorted DataFrame containing the top 10 rows with the lowest popularity values
sorted_df
```

# Describing Data Transpose : 

```python
# Generate descriptive statistics of the DataFrame df_tracks, transpose the result for better readability
df_tracks.describe().transpose()
```

# Most 10 Popular Songs :

```python
# Filter the DataFrame df_tracks to include only rows where the 'popularity' column is greater than 90
# Sort the filtered DataFrame by the 'popularity' column in descending order
most_popular = df_tracks.query('popularity > 90', inplace=False).sort_values('popularity', ascending=False)

# Select the top 10 rows from the sorted DataFrame
most_popular[:10]
```

# Checking artists Data

```python
# Select the 'artists' column for the 19th row (index 18) of the DataFrame df_tracks
df_tracks[["artists"]].iloc[18]
```

# Changing the duration from ms to s :

```python
# Calculate the duration of each track in seconds by dividing the duration in milliseconds by 1000 and rounding the result
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x / 1000))

# Drop the 'duration_ms' column from the DataFrame
# Specify inplace=True to modify the DataFrame in place, and axis=1 to indicate the column axis
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
# Display the first few rows of the 'duration' column in the DataFrame df_tracks
df_tracks.duration.head()
```

# Change Dates Format to Year: 

```python
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'], format='ISO8601')
```

# Correlation Heatmap Between Variables

```python
# Drop non-numeric columns from the DataFrame and store the result in numeric_df
numeric_df = df_tracks.drop(['id', 'name', 'artists', 'id_artists', 'key'], axis=1)

# Compute the correlation matrix of the numeric columns
corr_df = numeric_df.corr()

# Plot a heatmap of the correlation matrix using seaborn
plt.figure(figsize=(14, 6))  # Set the size of the figure
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")  # Plot heatmap
heatmap.set_title("Correlation Heatmap Between Variables")  # Set the title of the heatmap
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)  # Rotate x-axis labels for better readability
plt.show()  # Display the plot
```

# Distrubtion plot for Number of Songs per Year in Spotify:

```python
# Extract the year from the 'release_date' column and store it in the 'year' variable
year = df_tracks['release_date'].dt.year

# Create a discrete distribution plot using seaborn
# Pass the 'year' variable as the data to be plotted
# Set discrete=True to create a histogram-like plot
# Set aspect=2 to adjust the width-to-height ratio of the plot
# Set height=5 to adjust the height of the plot
# Use kind="hist" to specify the type of plot
# Set the title of the plot to "Number of Songs per Year"
sns.displot(year, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of Songs per Year")
```

# Bar Plot for Year vs Duration : 

```python
# Reset index of df_tracks
df_tracks.reset_index(drop=True, inplace=True)

# Extract year from the 'dates' column
year = df_tracks['release_date'].dt.year

# Define total duration
total_dr = df_tracks['duration_ms']

# Plot
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)  # Create a figure and axis with specified size
sns.barplot(x=year, y=total_dr, ax=ax)  # Create a bar plot
ax.set(title="Year vs Duration")  # Set title of the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()  # Show the plot
```

```python
# Reset index of df_tracks
df_tracks.reset_index(drop=True, inplace=True)

# Extract year from the 'dates' column
year = df_tracks['release_date'].dt.year

# Define total duration
total_dr = df_tracks['duration_ms']

# Calculate mean and standard deviation of total duration for each year
mean_duration = df_tracks.groupby(year)['duration_ms'].mean()
std_duration = df_tracks.groupby(year)['duration_ms'].std()

# Plot
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)  # Create a figure and axis with specified size
plt.errorbar(x=mean_duration.index, y=mean_duration, yerr=std_duration, fmt='-o', capsize=5)  # Create a line plot with error bars
ax.set(title="Year vs Mean Duration with Error Bars")  # Set title of the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines
plt.show()  # Show the plot
```

# Uploading New features file :

```python
# Read the Spotify features dataset from the CSV file into a DataFrame
df_genre = pd.read_csv('SpotifyFeatures.csv')

# Display the first few rows of the DataFrame to inspect the data
df_genre.head()
```

# BarPlot for Duration of Songs in Different Genres:

```python
# Set the title of the plot
plt.title("Duration of Songs in Different Genres")

# Create a bar plot showing the duration of songs in different genres with blue color
sns.barplot(y='genre', x='duration_ms', data=df_genre, palette='Blues')

# Set labels for x and y axes
plt.xlabel("Duration in Milliseconds")
plt.ylabel("Genres")

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

# BarPlot for Top 5 Genres by Popularity :

```python
# Set the style of the plot to 'darkgrid'
sns.set_style(style="darkgrid")

# Create a figure with specified size
plt.figure(figsize=(12,6))

# Sort the DataFrame by popularity in descending order and select the top 10 rows
famous = df_genre.sort_values("popularity", ascending=False).head(10)

# Create a bar plot showing the popularity of genres
sns.barplot(y='genre', x='popularity', data=famous)

# Set title of the plot
plt.title("Top 5 Genres by Popularity")

# Show the plot
plt.show()
```
